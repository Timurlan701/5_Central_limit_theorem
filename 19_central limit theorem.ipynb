Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

Generate 1 sample of volume 1000 from the [Beta distribution function](https://ru.wikipedia.org/wiki/Бета-распределение) with a = 0.6, b = 0.7

```python
a = 0.6
b = 0.7

beta_rv = sts.beta(a, b)
sample = beta_rv.rvs(1000)
x = np.linspace(0, 1, 1000)
```

Plot a sample histogram and probability density function

```python
plt.figure(figsize=(8,8))

plt.hist(sample, bins=50, density=True, label='sample pdf')
plt.plot(x, beta_rv.pdf(x), label='theoretical pdf')

plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.plot
```

Generate 1000 sample of volume 5, 10, 50, 100 from the Beta distribution

Plot histograms of distributions of their sample mean

```python
plt.figure(figsize=(8,8))

mu = a/(a+b)                    # theoretical EX Beta distribution function
sigma_sq = a*b/((a+b)**2*(a+b+1))  # theoretical DX Beta distribution fucttion

for n in 5, 10, 50, 100:
    sample_mean_hist = []
    for i in range(1000):
        sample = beta_rv.rvs(n)
        sample_mean = sum(sample)/n
        sample_mean_hist.append(sample_mean)
    
    plt.hist(sample_mean_hist, bins=50, alpha=0.5, density=True, label='sample mean n=%s' % n)
    norm_rv = sts.norm(loc = mu, scale = np.sqrt(sigma_sq/n))
    plt.plot(x, norm_rv.pdf(x), label='norm_pdf n=%s' % n)
    plt.legend()
    plt.ylabel('$f(x)$')
    plt.xlabel('$x$')
    plt.plot
print('mu =', round(mu, 4), 'sigma_sq = ', round(sigma_sq, 4))
```

As the sample size n increases, the distribution of the sample mean corresponds to the сentral limit theorem. 

As n increases, the accuracy of the normal approximation increases: the confidence interval narrows and the probability density function increases.
